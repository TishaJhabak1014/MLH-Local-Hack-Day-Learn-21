```python
from bs4 import BeautifulSoup 
import requests

with open('/content/drive/MyDrive/Colab_Notebooks/Python_projects/simple.html') as html_file: # passing the html file path
  soup = BeautifulSoup(html_file, 'lxml')   # passing the html file and the parser
```

```python
soup  # a beautifulsoup object of our parsed html
```

```python
print(soup.prettify())  # to get the parsed object output with proper indentation format
```

```python
# grabbing specific infomation associated with specific tags 
# grab title tag and text

title_tag = soup.title  # accessing tags by accessing it like an attribute
title_txt = soup.title.text

print(title_tag, title_txt, sep="\n")
```

```python
# to grab tags specifically, we use find() method

first_div = soup.div # grabs the first div with all the child tag
print(first_div)
print()
footer_div = soup.find('div', class_ = 'footer') 
print(footer_div)
```

```python
# accessing multiple things from an page
article = soup.find('div', class_= 'article')
print(article)

headline = article.h2.a.text
print("\n", headline)

summary = article.p.text
print('\n', summary)
```

```python
articles = soup.find_all('div', class_="article")
print(articles)  # a ResultSet() object(list like) of all divs with class article
for article in articles:
  print('\n', article.h2.a.text)
  print(article.p.text)
```

## Let us scrap with a motive. 

```python
from bs4 import BeautifulSoup
import requests
import csv

source = requests.get('http://coreyms.com').text  # extracting the html file from the response object

soup = BeautifulSoup(source, 'lxml')  

article = soup.find('article')
print(article.prettify())
headline = article.h2.a.text
print(headline)
summary = article.find('div', class_ = 'entry-content').p.text
print(summary)
link = article.find('iframe', class_= 'youtube-player')['src']
print(link)
link_id = link.split('/')[4].split('?')[0]
print(link_id)
yt_link = f'http://youtube.com/watch?v={link_id}'
print(yt_link)
```

```python
from bs4 import BeautifulSoup
import requests
import csv

source = requests.get('http://coreyms.com').text  # extracting the html file from the response object

soup = BeautifulSoup(source, 'lxml')  


csv_file = open('cms_scrape.csv', 'w') # open a csv file in write mode

csv_writer = csv.writer(csv_file)
csv_writer.writerow(['headline', 'summary', 'video_link'])  # headers of the csv file

for article in soup.find_all('article'): # looping over all the articles
    headline = article.h2.a.text
    print(headline)

    summary = article.find('div', class_='entry-content').p.text
    print(summary)

    try:
        vid_src = article.find('iframe', class_='youtube-player')['src']

        vid_id = vid_src.split('/')[4]
        vid_id = vid_id.split('?')[0]

        yt_link = f'https://youtube.com/watch?v={vid_id}'
    except Exception as e:
        yt_link = None

    print(yt_link)

    print()

    csv_writer.writerow([headline, summary, yt_link])  # saving the info. into a csv file

csv_file.close()
```
